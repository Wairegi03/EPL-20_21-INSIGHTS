```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
df=pd.read_csv('EPL_20_21.csv')
df.head()
```

```python
#to calculate the number of penalty missed:
penalty_goals=df['Penalty_Goals'].sum()
penalty_missed=df['Penalty_Attempted'].sum()-penalty_goals
print('TOTAL PENALTY MISSED IS :',penalty_missed)
```

```python
#piechart To show total penalty scored vs penalty missed:
plt.figure(figsize=(10,16))
data=[penalty_goals,pena_missed]
labels=['Penalty Scored','Penalty Missed']
color=sns.color_palette('Set1')
plt.pie(data,labels=labels,colors=color,autopct='%.0f%%')
plt.show()

```

```python
#total nationality represented in the english premier leaugue for that season;
np.size((df.Nationality.unique()))
```

```python
#10 most nations REPRESENTed IN A BAR CHART:
df.groupby('Nationality').size().sort_values(ascending=False).head(10).plot(kind='bar',color=sns.color_palette('magma'))
```

```python
#5 most clubs with most of the squad members available:
df['Club'].value_counts().nlargest(5).plot(kind='bar',color=sns.color_palette("viridis"))
```

```python
#different sets of age groups:
u20=df[df.Age <=20]
A20_25=df[(df.Age>20)&(df.Age<=25)]
A25_30=df[(df.Age>25)&(df.Age<=30)]
above30=df[df.Age>30]
```

```python
#pie chart representation of the above data:
x=np.array([u20['Name'].count(),A20_25['Name'].count(),A25_30['Name'].count(),above30['Name'].count()])
mylabels=['u20','A20_25','A25_30','above30']
plt.title('piechart for various age groups',fontsize=20)
plt.pie(x,labels=mylabels,autopct='%.1f%%')
plt.show()
```

```python
#total under_20 players in each club.
under_20=df[df.Age<=20]
under_20['Club'].value_counts().plot(kind='bar', color= sns.color_palette('cubehelix'))
```

```python
#average age in each club using a boxplot:
plt.figure(figsize=(16,15))
sns.boxplot(x='Club',y='Age',data=df) 
plt.xticks(rotation=90)
```

```python
#a list of top 10 top scorers:
df[['Club','Goals','Name']].nlargest(n=10,columns='Goals')
```

```python
#average age of each club:
num_of_players_each_club=df.groupby('Club').size()
average_age_per_club=df.groupby('Club')['Age'].sum() / num_of_players_each_club 
average_age_per_club.sort_values(ascending=False)
```

```python
#barchart for goals scored by each club:
df.groupby('Club',as_index=False)['Goals'].sum().nlargest(n=10,columns='Goals').plot(kind='bar',color=sns.color_palette('cubehelix'))
```
